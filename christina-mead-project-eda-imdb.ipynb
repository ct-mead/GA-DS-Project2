#### Determine the top rated movie (by star rating) for each genre.

```python
d = dict()
trs = movies.groupby('genre').star_rating.max()
for i in trs.index:
    g = movies[movies['genre'] == i]
    mr = trs[i]
    trm = g[g.star_rating == mr]
    d[i] = trm
    
for k in d.keys():
    m = d[k]
    t = m.title
    r = m.star_rating
    print(k, t.values[0], r.values[0])

#There must be a better way ðŸ˜±
```
